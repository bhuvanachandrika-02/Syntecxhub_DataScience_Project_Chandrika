```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Upload file (only for Google Colab)
from google.colab import files
uploaded = files.upload()

# Step 2: Load dataset
df = pd.read_csv('data.csv')

# Step 3: View data
print("First 5 rows:")
print(df.head())

print("\nLast 5 rows:")
print(df.tail())

# Step 4: Data info
print("\nData Info:")
print(df.info())

# Step 5: Summary statistics
print("\nSummary Statistics:")
print(df.describe())

# Step 6: Check missing values
print("\nMissing Values:")
print(df.isnull().sum())

# Step 7: Filtering (FIXED ✅ use numeric column)
filtered = df[df['Marks'] > 80]

print("\nStudents with Marks > 80:")
print(filtered)

# Step 8: Select columns
selected = df[['Name', 'Marks']]

print("\nSelected Columns (Name & Marks):")
print(selected)

# Step 9: Save filtered data
filtered.to_csv('filtered_output.csv', index=False)

# Step 10: Visualization
plt.bar(df['Name'], df['Marks'])
plt.xticks(rotation=45)
plt.title("Student Marks")
plt.xlabel("Name")
plt.ylabel("Marks")
plt.show()

print("\n✅ Task Completed Successfully!")
```
